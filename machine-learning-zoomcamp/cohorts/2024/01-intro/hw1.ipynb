```python
median_screen = df['Screen'].median()
print(f'{median_screen=}')

most_frequent_screen = df['Screen'].mode()[0]
print(f'{most_frequent_screen=}')

df['Screen'].fillna(most_frequent_screen, inplace=True)

new_median_screen = df['Screen'].median()
print(f'{new_median_screen=}')
```

```python
innjoo_df = df[df['Brand'] == 'Innjoo']
innjoo_df = innjoo_df[['RAM', 'Storage', 'Screen']]

X = innjoo_df.values

XTX = X.T.dot(X)

XTX_inv = np.linalg.inv(XTX)

y = np.array([1100, 1300, 800, 900, 1000, 1100])

XTX_inv_XT = XTX_inv.dot(X.T)
w = XTX_inv_XT.dot(y)
print(f'{w.sum()=}')
```
